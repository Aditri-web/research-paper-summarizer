```python
import requests
import json
from datasets import Dataset, DatasetDict

print("Downloading SciTLDR dataset directly...")

# Correct URLs - using the actual file structure from the allenai/scitldr repo
base_url = "https://huggingface.co/datasets/allenai/scitldr/resolve/main/SciTLDR-Data/SciTLDR-A/ctrl"

urls = {
    "train": f"{base_url}/train.jsonl",
    "test": f"{base_url}/test.jsonl",
    "validation": f"{base_url}/dev.jsonl"
}

def download_jsonl(url):
    """Download a JSONL file and return a list of records."""
    print(f"  Fetching: {url}")
    response = requests.get(url)
    if response.status_code != 200:
        raise ValueError(f"Failed to download {url} — Status code: {response.status_code}")
    records = []
    for line in response.text.strip().split("\n"):
        if line:
            records.append(json.loads(line))
    return records

# Download all splits
raw_data = {}
for split, url in urls.items():
    raw_data[split] = download_jsonl(url)
    print(f"  {split}: {len(raw_data[split])} samples loaded")

# Convert to HuggingFace DatasetDict
dataset = DatasetDict({
    split: Dataset.from_list(records)
    for split, records in raw_data.items()
})

print(f"\nDataset structure: {dataset}")
print(f"\nSplits available: {list(dataset.keys())}")
print(f"\nTrain samples:      {len(dataset['train'])}")
print(f"Validation samples: {len(dataset['validation'])}")
print(f"Test samples:       {len(dataset['test'])}")

# Preview a sample
print("\n--- Example Sample ---")
sample = dataset['train'][0]
print(f"Source: {' '.join(sample['source'][:2])}...")
print(f"Target: {sample['target'][0]}")
```

```python
import requests
import json
from datasets import Dataset, DatasetDict

print("Loading SciTLDR dataset from direct URLs...")

# Corrected URLs for the 'Abstract' (SciTLDR-A) configuration
base_url = "https://huggingface.co/datasets/allenai/scitldr/resolve/main/SciTLDR-Data/SciTLDR-A"

urls = {
    "train": f"{base_url}/train.jsonl",
    "test": f"{base_url}/test.jsonl",
    "validation": f"{base_url}/dev.jsonl"
}

def download_jsonl(url):
    """Download a JSONL file and parse it into a list of dictionaries."""
    print(f"  Downloading: {url}")
    response = requests.get(url)
    if response.status_code != 200:
        raise ValueError(f"Error {response.status_code}: Could not find file at {url}")
    
    # Parse each line as a JSON object
    records = [json.loads(line) for line in response.text.strip().split("\n") if line.strip()]
    return records

try:
    # 1. Download and Parse
    raw_splits = {}
    for split, url in urls.items():
        raw_splits[split] = download_jsonl(url)
        print(f"  Successfully loaded {len(raw_splits[split])} {split} samples.")

    # 2. Convert to Hugging Face Dataset object
    dataset = DatasetDict({
        split: Dataset.from_list(records)
        for split, records in raw_splits.items()
    })

    # 3. Print Results
    print(f"\nDataset structure: {dataset}")
    print(f"\nSplits available: {list(dataset.keys())}")
    print(f"\nTrain samples: {len(dataset['train'])}")
    print(f"Validation samples: {len(dataset['validation'])}")
    print(f"Test samples: {len(dataset['test'])}")

    # 4. Show a sample
    print("\n--- Example Sample (First Train Entry) ---")
    first_sample = dataset['train'][0]
    # 'source' is a list of sentences, 'target' is a list of summaries
    print(f"Source: {' '.join(first_sample['source'][:2])}...")
    print(f"TLDR (Target): {first_sample['target'][0]}")

except Exception as e:
    print(f"\nError: {e}")
```

```python
import requests
import json
from datasets import Dataset, DatasetDict

# Load SciTLDR dataset (small, perfect for starting)
print("Loading SciTLDR dataset (Manual Load Mode)...")

# These are the direct URLs to the data files on Hugging Face
base_url = "https://huggingface.co/datasets/allenai/scitldr/resolve/main/SciTLDR-Data/SciTLDR-A"
urls = {
    "train": f"{base_url}/train.jsonl",
    "test": f"{base_url}/test.jsonl",
    "validation": f"{base_url}/dev.jsonl"
}

def load_from_url(url):
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Failed to download {url}")
    # Parse JSONL format
    return [json.loads(line) for line in response.text.strip().split('\n') if line]

# Download and create the dataset object
try:
    dataset = DatasetDict({
        "train": Dataset.from_list(load_from_url(urls["train"])),
        "test": Dataset.from_list(load_from_url(urls["test"])),
        "validation": Dataset.from_list(load_from_url(urls["validation"]))
    })

    print(f"\nDataset structure: {dataset}")
    print(f"Splits available: {list(dataset.keys())}")
    print(f"Train samples: {len(dataset['train'])}")
    print(f"Validation samples: {len(dataset['validation'])}")
    print(f"Test samples: {len(dataset['test'])}")
    
    # Success Message
    print("\nDataset successfully loaded and ready for use!")

except Exception as e:
    print(f"Failed to load dataset: {e}")
```
